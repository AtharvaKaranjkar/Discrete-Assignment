```python
# ===============================================================
#  Counting binary strings of length n with NO run of "111"
#  Prints:
#      • Recurrence steps
#      • Final a_n
#      • ALL valid binary strings of length n
# ===============================================================

from itertools import product

# Take input
n = int(input("Enter value of n: "))

# --- Recurrence part -------------------------------------------------------- #

if n == 0:
    print("a_0 = 1")
elif n == 1:
    print("a_1 = 2")
elif n == 2:
    print("a_2 = 4")
else:
    a0, a1, a2 = 1, 2, 4
    print("Recurrence values:")
    print("a_0 =", a0)
    print("a_1 =", a1)
    print("a_2 =", a2)

    for i in range(3, n+1):
        a3 = a2 + a1 + a0
        print(f"a_{i} = {a3}")
        a0, a1, a2 = a1, a2, a3

    print("\nFinal Answer from recurrence:")
    print(f"a_{n} = {a3}")

# --- Brute force valid strings -------------------------------------------- #

print("\nValid binary strings of length", n, "(no '111'):\n")

valid_strings = []

for bits in product("01", repeat=n):
    s = "".join(bits)
    if "111" not in s:
        valid_strings.append(s)
        print(s)

print("\nTotal valid strings =", len(valid_strings))
```
